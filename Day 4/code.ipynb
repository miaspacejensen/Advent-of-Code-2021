```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n\n")
    return lines
```

```python
def get_final_score(data):
    nums_drawn = [int(val) for val in data[0].split(",")]

    boards = [[[int(val) for val in row.split()] for row in board.split("\n")] for board in data[1:]]
    board_dfs = []
    for board in boards:
        board_df = pd.DataFrame(board)
        board_dfs.append(board_df)

    bingo = False
    for num in nums_drawn:
        # print(num)
        for i, df in enumerate(board_dfs):
            board_dfs[i] = df = df.replace(num, 'X')
            rows_complete = df.index[df.eq('X').all(axis=1)].tolist()
            cols_complete = df.columns[df.eq('X').all()].tolist()
            if len(rows_complete) > 0 or len(cols_complete) > 0:
                bingo = True
                print("Bingo! Board", i+1, "wins")
                display(df)

                sum_unmarked = df.applymap(lambda x: x if isinstance(x, (int, float)) else 0).values.sum()
                final_score = num * sum_unmarked

                break
        if bingo == True:
            break

    return final_score
```

```python
data = get_data("input_sample.txt")
final_score = get_final_score(data)
final_score
```

```python
data = get_data("input.txt")
final_score = get_final_score(data)
final_score
```

```python
def get_final_score(data):
    nums_drawn = [int(val) for val in data[0].split(",")]

    boards = [[[int(val) for val in row.split()] for row in board.split("\n")] for board in data[1:]]
    board_dfs = []
    for board in boards:
        board_df = pd.DataFrame(board)
        board_dfs.append(board_df)

    winners = []
    num_boards = len(boards)
    for num in nums_drawn:
        # print(num)
        for i, df in enumerate(board_dfs):
            board_dfs[i] = df = df.replace(num, 'X')
            rows_complete = df.index[df.eq('X').all(axis=1)].tolist()
            cols_complete = df.columns[df.eq('X').all()].tolist()
            if len(rows_complete) > 0 or len(cols_complete) > 0:
                if f"Board {i+1}" not in winners:
                    print("Bingo! Board", i+1, "wins")
                    winners.append(f"Board {i+1}")
                    print("Winners:", winners)
                    display(df)
                    sum_unmarked = df.applymap(lambda x: x if isinstance(x, (int, float)) else 0).values.sum()
                    final_score = num * sum_unmarked
        if num_boards == len(winners):
            break
        
    return final_score
```
