```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    pairs = [line.split(" -> ") for line in lines]
    coord_pairs = [[[int(val) for val in coord.split(',')] for coord in pair] for pair in pairs]
    return coord_pairs
```

```python
def count_visits(coord_pairs):
    coords_visited = {}

    for coord1, coord2 in coord_pairs:

        x1, y1 = coord1
        x2, y2 = coord2

        # same row
        if y1 == y2: 
            y = y1
            coord_set = {(x1, y1), (x2, y2)}
            if x1 < x2:
                start = x1 
                end = x2 
            elif x1 > x2:
                start = x2 
                end = x1
            for i in range(start, end):
                new_coord = (i, y)
                coord_set.add(new_coord)

        # same column
        elif x1 == x2: 
            x = x1
            coord_set = {(x1, y1), (x2, y2)}
            if y1 < y2:
                start = y1 
                end = y2 
            elif y1 > y2:
                start = y2 
                end = y1
            for i in range(y1, y2):
                new_coord = (x, i)
                coord_set.add(new_coord)

        # diagonal
        else: 
            coord_set = {}

        # count occurrances
        for coord in coord_set:
            if coord not in coords_visited.keys():
                coords_visited[coord] = 1
            else:
                coords_visited[coord] += 1
                
    return coords_visited
```

```python
def display_grid(coords_visited):
    
    n_cols = max([coord[0] for coord in list(coords_visited.keys())])+1
    nrows = max([coord[1] for coord in list(coords_visited.keys())])+1

    grid = [['.' for _ in range(n_cols)] for _ in range(nrows)]

    for coord in list(coords_visited.keys()):
        col_index, row_index = coord
        grid[row_index][col_index] = str(coords_visited[coord])

    for row in grid:
        print("".join(row))
```

```python
file_name = "input_sample.txt"
coord_pairs = get_data(file_name)
coords_visited = count_visits(coord_pairs)
display_grid(coords_visited)
count = len({key: value for key, value in coords_visited.items() if value >= 2})
print("\nCount =", count)
```
